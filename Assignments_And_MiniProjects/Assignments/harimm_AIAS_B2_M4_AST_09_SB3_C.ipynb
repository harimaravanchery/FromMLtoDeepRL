## Introduction

Stable Baselines3 (SB3) is a powerful and robust Python library for implementing reinforcement learning (RL) algorithms. It provides a collection of high-quality, well-maintained implementations of popular RL algorithms built on top of PyTorch. Here's a breakdown of SB3's key features and benefits:

**Features:**

* **Wide Range of RL Algorithms:** SB3 offers diverse RL algorithms like Proximal Policy Optimization (PPO), Soft Actor-Critic (SAC), Deep Q-Network (DQN), Twin Delayed Deep Deterministic Policy Gradient (TD3), and many more. This allows you to choose the best algorithm for your specific task and problem domain.
* **Modular and Flexible:** SB3 is designed with modularity in mind. Each component, like policies, environments, and replay buffers, is independent and can be easily swapped or customized. This flexibility enables you to tailor the library to your specific needs and research endeavors.
* **High Performance and Efficiency:** The PyTorch backend ensures efficient computation and fast training. SB3 utilizes various optimization techniques like automatic batching and vectorized operations to maximize performance.
* **Easy to Use and Get Started:** SB3 comes with comprehensive documentation, tutorials, and examples, making it accessible even for beginners in RL. The clear API and code structure facilitate quick learning and experimentation.
* **Active Community and Support:** SB3 has a vibrant community of researchers and developers. You can find extensive online resources, discussions, and helpful contributions on platforms like GitHub and forums.




**Pros of Stable Baselines3 (SB3):**

* **Wide Range of RL Algorithms:** SB3 offers a diverse selection of well-implemented and maintained RL algorithms like PPO, SAC, DQN, TD3, and many more. This allows you to choose the best algorithm for your specific task and problem domain.

## Example: Bipedal Walker

Here, the task is to teach the bipedal walker, to walk without crashing.

<br>

This is a simple 4-joint walker robot environment. There are two versions:

- Normal, with slightly uneven terrain
- Hardcore, with ladders, stumps, pitfalls

To solve the normal version, you need to get 300 points in 1600 time steps. To solve the hardcore version, you need 300 points in 2000 time steps.

<br>

Let's see the details of different aspects, such as rewards, states, and actions, that needs to be considered while modeling an RL solution for this problem.


### **Actions**

The action is a 4-dimensional vector denoting motor speed values in the [-1, 1] range for each of the 4 joints at both hips and knees.

### **States** (Observations)

The state is an 24-dimensional vector consisting of:
- hull angle speed
- angular velocity
- horizontal speed
- vertical speed
- position of joints (4)
- joints angular speed (4)
- legs contact with ground (2)
- 10 lidar rangefinder measurements (10)

There are no coordinates in the state vector.

### **Rewards**

Reward is given for moving forward, totaling 300+ points up to the far end. If the robot falls, it gets -100. Applying motor torque costs a small amount of points. A more optimal agent will get a better score.

### **Starting State**

The walker starts standing at the left end of the terrain with the hull horizontal, and both legs in the same position with a slight knee angle.

### **Episode End**

The episode will terminate
- if the hull gets in contact with the ground, or
- if the walker exceeds the right end of the terrain length.

To know more about the BipedalWalker environment, refer [here](https://gymnasium.farama.org/environments/box2d/bipedal_walker/).

```python
models_dir = "models/PPO"
logdir = "logs"

if not os.path.exists(models_dir):
    os.makedirs(models_dir)

if not os.path.exists(logdir):
    os.makedirs(logdir)

env = gym.make('BipedalWalker-v3')
env.reset()

model = PPO('MlpPolicy', env, verbose=1, tensorboard_log=logdir)

TIMESTEPS = 10000

for i in range(1,30):
    model.learn(total_timesteps=TIMESTEPS, reset_num_timesteps=False, tb_log_name="PPO")
    model.save(f"{models_dir}/{TIMESTEPS*i}")
```

### Proximal Policy Optimization (PPO)





* **Policy Gradient Algorithm:** PPO belongs to the family of policy gradient algorithms, directly optimizing the policy (the agent's decision-making process) to maximize expected rewards.
* **Trust Region Approach:** It employs a clipped surrogate objective function to limit policy updates within a trust region, preventing overly large updates that could destabilize training.
* **On-Policy Algorithm:** It learns directly from experiences generated by the current policy, often leading to more sample-efficient learning compared to off-policy algorithms.
* **Actor-Critic Architecture:** PPO utilizes an actor-critic framework:
    - **Actor:** A neural network that outputs actions based on the current state.
    - **Critic:** A neural network that estimates the value of state-action pairs, guiding the actor's updates.

**Key Steps in SB3's PPO Implementation:**

1. **Collect Experiences:** The agent interacts with the environment, collecting experiences (state, action, reward, next state) tuples.
2. **Calculate Advantage Estimates:** The critic network estimates the advantages of actions taken in each state, indicating how much better those actions were compared to expected performance.
3. **Update the Actor:** The actor network is updated using the clipped surrogate objective function, aiming to increase the probability of actions with high estimated advantages.
4. **Update the Critic:** The critic network is updated to improve its value estimations, using a standard temporal difference (TD) learning approach.
5. **Repeat:** The process iterates, collecting more experiences and updating networks until a satisfactory policy is learned.

**Advantages of PPO:**

* **Sample Efficiency:** Often learns effectively with fewer experiences compared to other policy gradient algorithms.
* **Stability:** The trust region approach helps prevent instability and divergence during training.
* **Relatively Easy to Implement and Tune:** Considered more straightforward to implement and tune than some other RL algorithms.
* **Good Performance Across Diverse Tasks:** Has demonstrated successful results in various continuous control tasks, including robotics, games, and more.

**Considerations:**

* **Hyperparameter Sensitivity:** Can be sensitive to hyperparameter choices, requiring careful tuning.
* **Computational Cost:** Can be computationally expensive, especially for complex tasks and large neural networks.

**Overall, PPO is a powerful and popular RL algorithm known for its stability, sample efficiency, and performance in various continuous control domains. Stable Baselines3 provides a well-optimized and user-friendly implementation of PPO, making it accessible for experimentation and research.**

### A2C




* **Actor-Critic Architecture:** Similar to PPO, it utilizes an actor network for action selection and a critic network for state-action value estimation.
* **On-Policy Algorithm:** Directly learns from experiences gathered by the current policy, often leading to sample-efficient learning.
* **Advantage Actor-Critic:** Updates the actor based on advantage estimates, which indicate how much better actions were compared to expected performance.
* **Asynchronous Execution:** Multiple actors (workers) simultaneously collect experiences and update the policy in parallel, potentially speeding up training.

**Key Steps in SB3's A2C Implementation:**

1. **Independent Actors:** Multiple actor workers interact with the environment, gathering experiences.
2. **Advantage Estimation:** Each worker calculates advantage estimates from their collected experiences.
3. **Gradient Updates:** Each worker performs a gradient update on the shared global actor and critic networks using its local data and advantage estimates.
4. **Synchronization:** Periodically, the global networks are synchronized across all workers.

**Advantages of A2C:**

* **Potential Speedup:** Asynchronous execution can speed up training compared to synchronous algorithms like PPO, especially for tasks with long environment steps.
* **Sample Efficiency:** On-policy learning and advantage estimates often lead to efficient learning.
* **Simpler Implementation:** Generally easier to implement than complex algorithms like PPO or TD3.

**Considerations:**

* **Potential Instability:** Asynchronous updates can introduce instability in some setups, requiring careful parameter tuning.
* **Synchronization Overhead:** Frequent synchronization across workers can incur some overhead, slightly reducing speedup potential.
* **Limited Multi-Agent Applications:** Primarily suited for single-agent environments, as extending it to multi-agent settings can be challenging.

**Overall, A2C in SB3 provides a potentially faster and simpler alternative to PPO for single-agent reinforcement learning tasks. However, consider its potential for instability and limitations in multi-agent settings before choosing it for your project.**
